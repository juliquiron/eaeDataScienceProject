```python
# Importamos librer√≠as
import sklearn.datasets as datasets
import pandas as pd
import pydotplus
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
# Cargamos los datos
iris = datasets.load_iris()
# Definimos un Dataframe (estructura tabular) con las variables predictoras
# y por otro lado un vector apartado con la variable respuesta
df = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
df.head()
```

```python
# Partir datos en train y test
X_train, X_test, y_train, y_test = train_test_split(
df, y, test_size=0.25, random_state=42)
# Crear modelo
dtree = DecisionTreeClassifier()
# Entrenar modelo
dtree.fit(X_train, y_train)
#########
# Gr√°fico
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,
filled=True, rounded=True,
special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
#########
```

```python
X_train, X_test, y_train, y_test = train_test_split(
df, y, test_size=0.25, random_state=4)
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
#########
# Gr√°fico
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
#########
```
